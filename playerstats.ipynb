```python
import json
import pandas as pd
```

```python
puuid = "4WU60VKd11pKuuSBe2lnGekDpuOY3b_uzQIGNzS6uuo7ymRmmNlfZ7DIAYSfhkUmlqOkE1jbfxPJRQ" # this is for kungpowji
with open("matches.json", "r") as f:
    matches = json.load(f)

records = []
for match in matches:
    if not match or "info" not in match:
        continue
    info = match["info"]
    for p in info["participants"]:
        if p["puuid"] == puuid:  # use your puuid
            records.append({
                "matchId": match["metadata"]["matchId"],
                "timestamp": info["gameCreation"],
                "duration": info["gameDuration"],
                "champion": p["championName"],
                "role": p["teamPosition"],
                "win": p["win"],
                "kills": p["kills"],
                "deaths": p["deaths"],
                "assists": p["assists"],
                "cs": p["totalMinionsKilled"] + p["neutralMinionsKilled"],
                "gold": p["goldEarned"],
                "vision": p["visionScore"],
                "damage": p["totalDamageDealtToChampions"]
            })

kungpow_rec = pd.DataFrame(records)

```

```python
records = []
for match in matches:
    if "info" not in match:
        continue

    info = match["info"]
    game_duration_minutes = info["gameDuration"] / 60
    timestamp = pd.to_datetime(info["gameCreation"], unit="ms")

    for p in info["participants"]:
        if p["puuid"] == puuid:  # only keep your games
            cs = p.get("totalMinionsKilled", 0) + p.get("neutralMinionsKilled", 0)
            cs_per_min = cs / game_duration_minutes if game_duration_minutes > 0 else 0
            records.append({
                "date": timestamp,
                "champion": p["championName"],
                "csPerMin": cs_per_min
            })

df = pd.DataFrame(records).sort_values("date")
print(df.head())
```
